```python
chile = data[(data.country == "Chile")]
china = data[(data.country == "China")]
germany = data[(data.country == "Germany")]
mexico = data[(data.country == "Mexico")]
united =data[(data.country == "United States of America")]
zimbabwe =data[(data.country == "Zimbabwe")]
```

```python
sns.lmplot(x="gdp", y = "life expectancy", data = data, hue= "country", fit_reg=False, palette = "flare")
plt.show()
```
